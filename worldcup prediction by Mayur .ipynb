```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# bowling_card = pd.read_csv('bowling_card.csv')
# batting_card = pd.read_csv('batting_card.csv')

# Check if 'year' column exists, if not, add it (Example: using a 'match_date' column if available)
if 'year' not in bowling_card.columns:
    if 'date' in bowling_card.columns:
        bowling_card['year'] = pd.to_datetime(bowling_card['date']).dt.year
    else:
        # For illustration, if 'date' is not present, we assume a random year (replace with actual logic)
        bowling_card['year'] = pd.Series([2020] * len(bowling_card))  # Replace with actual logic to extract or assign the year

if 'year' not in batting_card.columns:
    if 'date' in batting_card.columns:
        batting_card['year'] = pd.to_datetime(batting_card['date']).dt.year
    else:
        # For illustration, if 'date' is not present, we assume a random year (replace with actual logic)
        batting_card['year'] = pd.Series([2020] * len(batting_card))  # Replace with actual logic to extract or assign the year

### Bowling Performance Analysis Over Years ###

# Calculating Economy Rate
bowling_card['economy_rate'] = bowling_card['conceded'] / bowling_card['overs']

# Calculating Average
bowling_card['average'] = bowling_card['conceded'] / bowling_card['wickets']

# Calculating Strike Rate
bowling_card['strike_rate'] = bowling_card.apply(lambda row: row['overs'] / row['wickets'] if row['wickets'] > 0 else None, axis=1)

# Aggregating yearly data for bowlers
bowling_yearly = bowling_card.groupby(['fullName', 'year']).agg({
    'conceded': 'sum',
    'wickets': 'sum',
    'overs': 'sum',
    'maidens': 'sum',
    'dots': 'sum',
    'foursConceded': 'sum',
    'sixesConceded': 'sum',
    'wides': 'sum',
    'noballs': 'sum'
}).reset_index()

# Recalculate metrics after yearly aggregation
bowling_yearly['economy_rate'] = bowling_yearly['conceded'] / bowling_yearly['overs']
bowling_yearly['average'] = bowling_yearly['conceded'] / bowling_yearly['wickets']
bowling_yearly['strike_rate'] = bowling_yearly.apply(lambda row: row['overs'] / row['wickets'] if row['wickets'] > 0 else None, axis=1)

### Batting Performance Analysis Over Years ###

# Calculating Batting Average
batting_card['batting_average'] = batting_card['runs'] / (batting_card['isNotOut'] + 1)

# Calculating Strike Rate
batting_card['strike_rate'] = (batting_card['runs'] / batting_card['ballsFaced']) * 100

# Calculating Fours and Sixes
batting_card['boundary_count'] = batting_card['fours'] + batting_card['sixes']

# Aggregating yearly data for batsmen
batting_yearly = batting_card.groupby(['fullName', 'year']).agg({
    'runs': 'sum',
    'isNotOut': 'sum',
    'ballsFaced': 'sum',
    'fours': 'sum',
    'sixes': 'sum'
}).reset_index()

# Recalculate metrics after yearly aggregation
batting_yearly['batting_average'] = batting_yearly['runs'] / (batting_yearly['isNotOut'] + 1)
batting_yearly['strike_rate'] = (batting_yearly['runs'] / batting_yearly['ballsFaced']) * 100
batting_yearly['boundary_count'] = batting_yearly['fours'] + batting_yearly['sixes']

### Merging Batting and Bowling Yearly Metrics ###

# Merge Batting and Bowling Yearly Metrics
full_yearly_metrics = pd.merge(batting_yearly, bowling_yearly, on=['fullName', 'year'], how='outer', suffixes=('_bat', '_bowl'))
full_yearly_metrics.fillna(0, inplace=True)  # Fill NA values with 0

### Visualizations ###

# Line plot for a specific player over years
def plot_player_performance(player_name):
    player_data = full_yearly_metrics[full_yearly_metrics['fullName'] == player_name]
    
    fig, ax1 = plt.subplots(figsize=(14, 8))

    ax1.set_title(f'Performance of {player_name} from 2008 to 2022')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Batting Metrics', color='tab:blue')

    ax1.plot(player_data['year'], player_data['batting_average'], label='Batting Average', color='tab:blue')
    ax1.plot(player_data['year'], player_data['strike_rate_bat'], label='Batting Strike Rate', color='tab:cyan')
    ax1.legend(loc='upper left')
    
    ax2 = ax1.twinx()  
    ax2.set_ylabel('Bowling Metrics', color='tab:green')
    ax2.plot(player_data['year'], player_data['economy_rate'], label='Economy Rate', color='tab:green')
    ax2.plot(player_data['year'], player_data['strike_rate_bowl'], label='Bowling Strike Rate', color='tab:olive')
    ax2.legend(loc='upper right')
    
    fig.tight_layout()
    plt.show()

# Example: Plot performance of a specific player
plot_player_performance('Player Name')  # Replace 'Player Name' with the actual player name

# Consistency visualization for all players (Standard Deviation over years)
consistency_metrics = full_yearly_metrics.groupby('fullName').agg({
    'batting_average': 'std',
    'strike_rate_bat': 'std',
    'economy_rate': 'std',
    'strike_rate_bowl': 'std'
}).reset_index()

consistency_metrics.rename(columns={
    'batting_average': 'batting_avg_std',
    'strike_rate_bat': 'bat_strike_rate_std',
    'economy_rate': 'econ_rate_std',
    'strike_rate_bowl': 'bowl_strike_rate_std'
}, inplace=True)

# Plot consistency
plt.figure(figsize=(14, 8))
sns.barplot(x='batting_avg_std', y='fullName', data=consistency_metrics.sort_values('batting_avg_std'), palette='viridis')
plt.title('Consistency of Batting Average (Lower is better)')
plt.xlabel('Standard Deviation of Batting Average')
plt.ylabel('Player Name')
plt.show()

plt.figure(figsize=(14, 8))
sns.barplot(x='econ_rate_std', y='fullName', data=consistency_metrics.sort_values('econ_rate_std'), palette='viridis')
plt.title('Consistency of Economy Rate (Lower is better)')
plt.xlabel('Standard Deviation of Economy Rate')
plt.ylabel('Player Name')
plt.show()

# You can add more visualizations for other metrics as needed
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# batting_card = pd.read_csv('batting_card.csv')

### Step 1: Calculate Ducks and Dismissals ###

# Counting the number of ducks (0 runs scored)
batting_card['ducks'] = batting_card['runs'].apply(lambda x: 1 if x == 0 else 0)

# Counting the number of dismissals (times a player is out)
batting_card['dismissals'] = batting_card['isNotOut'].apply(lambda x: 0 if x == 1 else 1)

# Aggregating the data to get the total ducks and dismissals for each player
player_misses = batting_card.groupby('fullName').agg({
    'ducks': 'sum',
    'dismissals': 'sum'
}).reset_index()

### Step 2: Visualizations ###

# Players with the most ducks
top_ducks = player_misses.sort_values('ducks', ascending=False).head(10)

plt.figure(figsize=(14, 8))
sns.barplot(x='ducks', y='fullName', data=top_ducks, palette='viridis')
plt.title('Top 10 Players with the Most Ducks in T20 World Cup')
plt.xlabel('Number of Ducks')
plt.ylabel('Player Name')
plt.show()

# Players with the most dismissals
top_dismissals = player_misses.sort_values('dismissals', ascending=False).head(10)

plt.figure(figsize=(14, 8))
sns.barplot(x='dismissals', y='fullName', data=top_dismissals, palette='viridis')
plt.title('Top 10 Players with the Most Dismissals in T20 World Cup')
plt.xlabel('Number of Dismissals')
plt.ylabel('Player Name')
plt.show()

### Step 3: Combined Analysis ###

# Players with the most ducks and dismissals
top_misses = player_misses.copy()
top_misses['total_misses'] = top_misses['ducks'] + top_misses['dismissals']
top_combined = top_misses.sort_values('total_misses', ascending=False).head(10)

plt.figure(figsize=(14, 8))
sns.barplot(x='total_misses', y='fullName', data=top_combined, palette='viridis')
plt.title('Top 10 Players with the Most Ducks and Dismissals Combined in T20 World Cup')
plt.xlabel('Total Misses (Ducks + Dismissals)')
plt.ylabel('Player Name')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace these with actual data loading)
# batting_card = pd.read_csv('batting_data.csv')
# bowling_card = pd.read_csv('bowling_data.csv')

# Calculate Batting Metrics
def calculate_batting_metrics(df):
    df['batting_average'] = df['runs'] / (df['isNotOut'] + 1)
    df['strike_rate'] = (df['runs'] / df['ballsFaced']) * 100
    return df.groupby('fullName').agg({
        'runs': 'sum',
        'batting_average': 'mean',
        'strike_rate': 'mean'
    }).reset_index()

batting_metrics = calculate_batting_metrics(batting_card)

# Calculate Bowling Metrics
def calculate_bowling_metrics(df):
    df['economy_rate'] = df['conceded'] / df['overs']
    df['bowling_average'] = df.apply(lambda row: row['conceded'] / row['wickets'] if row['wickets'] > 0 else None, axis=1)
    df['strike_rate'] = df.apply(lambda row: row['overs'] / row['wickets'] if row['wickets'] > 0 else None, axis=1)
    return df.groupby('fullName').agg({
        'wickets': 'sum',
        'bowling_average': 'mean',
        'economy_rate': 'mean',
        'strike_rate': 'mean'
    }).reset_index()

bowling_metrics = calculate_bowling_metrics(bowling_card)

# Calculate Team Performance Metrics
def calculate_team_performance(df):
    try:
        df['year'] = pd.to_datetime(df['date']).dt.year
    except KeyError:
        print("Error: 'date' column not found in the dataset.")
        return None
    
    team_performance = df.groupby('team').agg({
        'runs': 'sum',
        'year': 'nunique'
    }).reset_index()
    team_performance['matches'] = team_performance['year'] * 10  # Assuming 10 matches per year for simplification
    team_performance['win_rate'] = team_performance['runs'] / team_performance['matches']
    return team_performance

team_performance = calculate_team_performance(batting_card)

# Visualizations
sns.set(style="whitegrid")

# Top 10 Batsmen by Runs
top_batsmen = batting_metrics.sort_values('runs', ascending=False).head(10)
plt.figure(figsize=(14, 8))
sns.barplot(x='runs', y='fullName', data=top_batsmen, palette='viridis')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Total Runs')
plt.ylabel('Player Name')
plt.show()

# Top 10 Bowlers by Wickets
top_bowlers = bowling_metrics.sort_values('wickets', ascending=False).head(10)
plt.figure(figsize=(14, 8))
sns.barplot(x='wickets', y='fullName', data=top_bowlers, palette='viridis')
plt.title('Top 10 Bowlers by Total Wickets')
plt.xlabel('Total Wickets')
plt.ylabel('Player Name')
plt.show()

# Top 10 Players by Batting Strike Rate
top_strike_rate_batsmen = batting_metrics.sort_values('strike_rate', ascending=False).head(10)
plt.figure(figsize=(14, 8))
sns.barplot(x='strike_rate', y='fullName', data=top_strike_rate_batsmen, palette='viridis')
plt.title('Top 10 Batsmen by Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Player Name')
plt.show()

# Top 10 Teams by Win Rate
if team_performance is not None:
    plt.figure(figsize=(14, 8))
    sns.barplot(x='win_rate', y='team', data=team_performance, palette='viridis')
    plt.title('Top 10 Teams by Win Rate')
    plt.xlabel('Win Rate')
    plt.ylabel('Team Name')
    plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant performance metrics for clustering
selected_metrics = ['batting_average', 'strike_rate']

# Prepare data by selecting features
data = batting_card[selected_metrics].values

# Impute missing values with the mean of each feature column
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(data)

# Normalize the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(imputed_data)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Choose the number of clusters
kmeans.fit(normalized_data)

# Add cluster labels to the batting_card DataFrame
batting_card['cluster'] = kmeans.labels_

# Analyze clusters to find the best players over time
cluster_means = batting_card.groupby('cluster')[selected_metrics].mean()

# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='batting_average', y='strike_rate', hue='cluster', data=batting_card, palette='viridis')
plt.title('K-Means Clustering of Players')
plt.xlabel('Batting Average')
plt.ylabel('Strike Rate')
plt.legend(title='Cluster')
plt.show()

# Find the best players in each cluster
best_players = []
for cluster_label, group in batting_card.groupby('cluster'):
    best_player_idx = group['batting_average'].idxmax()  # Get the index of the player with the highest batting average
    best_player_info = group.loc[best_player_idx]  # Get the information of the best player
    best_players.append(best_player_info)

# Display information about the best players
for player_info in best_players:
    print(f"Best Player: {player_info['fullName']}")
    print(f"Country: {player_info['country']}")
    print(f"Best Cluster: {player_info['cluster']}\n")
```
